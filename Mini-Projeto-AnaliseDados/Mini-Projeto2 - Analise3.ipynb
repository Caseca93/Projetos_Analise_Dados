```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
df.head()
```

```python
df.fuelType
```

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
sns.set_style('white')
plt.subplots(figsize=(10,6))
sns.barplot(x='fuelType', y='price', data=df, hue='gearbox', palette='magma_r')
plt.xlabel('Tipo de Combustível', fontsize=20)
plt.ylabel('Preço Médio',fontsize=20)
plt.title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio', fontsize=12)
plt.savefig('plots/Veículos por tipo de combustível')
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
sns.set_style('white')
plt.subplots(figsize=(8,6))
sns.barplot(x='vehicleType', y='powerPS', data=df, hue='gearbox', palette='magma_')
plt.xlabel('Tipo de Veículo', fontsize=20)
plt.ylabel('Potência Média',fontsize=20)
plt.title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio', fontsize=12)
```
