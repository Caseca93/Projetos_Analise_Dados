## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
plt.subplots(figsize=(8,6))
sns.distplot(df['yearOfRegistration'], kde=True, color='#bcbd22')
plt.xlabel('Ano de Registro', fontsize=20)
plt.ylabel('Kde', fontsize=20)
plt.title('Distribuição de Veículos com base no Ano de Registro', fontsize=20)
plt.savefig('plots/Véiculos Distribuição')
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(df['vehicleType'], df['price'])
plt.ylabel('Preço', fontsize=20)
plt.xlabel('Tipos de Veículos', fontsize=20)
plt.text(5,27000,'Análise Outliers', fontsize=18, color='b')
plt.savefig("plots/analise_outliers")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
sns.set_style('darkgrid')
g = sns.catplot(x="vehicleType",data=df,kind='count', palette="winter", height=8, aspect=1)
plt.title('Veículos a venda conforme o tipo', fontsize=20)
plt.ylabel('Contagem', fontsize=20)
plt.xlabel('Tipo de Veículos', fontsize=20)
```

```python
# Salvando o plot
g.savefig("plots/contagem_veiculos_conforme_tipo")
```
