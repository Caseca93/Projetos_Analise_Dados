```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

```python
df.brand.value_counts()
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style('darkgrid')
sns.catplot(y='brand', data=df, kind='count', palette='winter_r', height=8, aspect=1.5)
plt.text(30000, 22, 'Número de veículos por cada marca', fontsize=20, color='b')
plt.xlabel('Números de Veículos', fontsize=15)
plt.ylabel('Marcas', fontsize=15)
plt.savefig('plots/Numero de veiculos por cada marca')
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
sns.set_style('whitegrid')
plt.subplots(figsize=(8,10))
sns.barplot(x='vehicleType', y='price', hue='gearbox', data=df, palette='pastel')
plt.title('Preço dos veículos por câmbio', fontsize=20)
plt.xlabel('Tipo de Veículos', fontsize=20)
plt.ylabel('Preço', fontsize=20)
```

```python
# Salvando o plot
plt.savefig('plots/veiculos por cambio preço')
```
